```python
import tkinter as tk
from tkinter import ttk

class MainApplication(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Wal-Mart Recipes")
        self.geometry("400x300")
       
        # Set up styles
        style = ttk.Style(self)
        style.configure("TButton", font=("Arial", 14, "bold"), foreground="#000000", background="#E0E0E0")
        style.map("TButton", background=[("active", "#D3D3D3")])
        style.configure("TLabel", font=("Arial", 18, "bold"), foreground="#333333")
        
      
        # Create buttons for different recipes
        recipe1_button = ttk.Button(self, text="Spaghetti with Meat Sauce", command=self.open_recipe1)
        recipe1_button.pack(pady=10)
        recipe2_button = ttk.Button(self, text="Hand Breaded Fried PorkChops", command=self.open_recipe2)
        recipe2_button.pack(pady=10)
        recipe3_button = ttk.Button(self, text="Chicken Alfredo", command=self.open_recipe3)
        recipe3_button.pack(pady=10)
        
    


    def open_recipe1(self):
        # New window for Spaghetti recipe
        recipe1_window = Recipe1Window(self)

    def open_recipe2(self):
        # New Window for PorkChops recipe
        recipe2_window = Recipe2Window(self)
        
    def open_recipe3(self):
        # New window for Chicken Alfredo
        recipe3_window = Recipe3Window(self)

class Recipe1Window(tk.Toplevel):
    def __init__(self, master):
        super().__init__(master)
        self.title("Spaghetti with Meat Sauce")
        self.geometry("400x300")

        # Set up styles
        style = ttk.Style(self)
        style.configure("TLabel", font=("Arial", 18, "bold"), foreground="#333333")
        style.configure("TFrame", background="#808080")

        # Create a frame to hold content
        content_frame = ttk.Frame(self)
        content_frame.pack(expand=True, fill="both", padx=20, pady=20)

        # Add content for Spaghetti
        title_label = ttk.Label(content_frame, text="Spaghetti")
        title_label.pack(pady=10)
        ingredients_label = ttk.Label(content_frame, text="Ingredients:")
        ingredients_label.pack(pady=10)
        ingredients_text = tk.Text(content_frame, height=5, width=40)
        ingredients_text.insert(tk.END, "1. Brown One pound of Ground Beef\n2. Drain/ Add One Jar of Spaghetti Sauce\n3. One Tablespoon Of sugar/ Add Garlic Powder to taste\n4. In a Seperate Pan Boil 6 Cups of water\n5. Add Angel Hair Pasta Cook and Drain")
        ingredients_text.configure(state="disabled")
        ingredients_text.pack(pady=10)

class Recipe2Window(tk.Toplevel):
    def __init__(self, master):
        super().__init__(master)
        self.title("Hand Breaded Pork Chops")
        self.geometry("400x300")

        # Set up styles
        style = ttk.Style(self)
        style.configure("TLabel", font=("Arial", 18, "bold"), foreground="#333333")
        style.configure("TFrame", background="#808080")

        # Create a frame to hold content
        content_frame = ttk.Frame(self)
        content_frame.pack(expand=True, fill="both", padx=20, pady=20)

        # Add content for PorkChops
        
        title_label = ttk.Label(content_frame, text="Pork Chops")
        title_label.pack(pady=10)
        ingredients_label = ttk.Label(content_frame, text="Ingredients:")
        ingredients_label.pack(pady=10)
        ingredients_text = tk.Text(content_frame, height=5, width=40)
        ingredients_text.insert(tk.END, ("1.Breading \n2. One Cup of All Purpose Flour\n 3 Cups Of Frosted Corn Flakes/n 2 Tablespoons of Lowrys All purpose Seasoning\n 2 Tablespoons of Lemon Pepper"))
        ingredients_text.insert(tk.END, ("Scramble two eggs in a bowl/n 2.Dip Pork Chops into eggs then breading\n3. Fry in Vegetable oil or air frier until Internal Temperature reaches 160 Degrees"))
        ingredients_text.configure(state="disabled")
        ingredients_text.pack(pady=10)
    

class Recipe3Window(tk.Toplevel):
    def __init__(self, master):
        super().__init__(master)
        self.title("Chicken Fettuccini Alfredo")
        self.geometry("400x300")
        
   
   # Set up styles
        style = ttk.Style(self)
        style.configure("TLabel", font=("Arial", 18, "bold"), foreground="#333333")
        style.configure("TFrame", background="#808080")

    # Create a frame to hold content
        content_frame = ttk.Frame(self)
        content_frame.pack(expand=True, fill="both", padx=20, pady=20)

    # Add content for Chicken Alfredo
        title_label = ttk.Label(content_frame, text="Chicken Fettuccini Alfredo")
        title_label.pack(pady=10)
        ingredients_label = ttk.Label(content_frame, text="Ingredients:")
        ingredients_label.pack(pady=10)
        ingredients_text = tk.Text(content_frame, height=5, width=40)
        ingredients_text.insert(tk.END, ("1.Cook Boneless Skinless Chicken Breasts and Set aside\n2. Heat One Jar of Alfredo Sauce Add Garlic Powder to taste\n3. In a Seperate Pan Boil 6 Cups of water Add Fetticini Pasta Cook until tender\n4. Drain Pasta and Mix"))
        ingredients_text.configure(state="disabled")
        ingredients_text.pack(pady=10)
    
if __name__ == "__main__":
    app = MainApplication()
    app.mainloop()
```
